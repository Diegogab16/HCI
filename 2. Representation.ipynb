### TSNE

```python

indices = np.random.choice(user.shape[0], 568, replace = False)
X= user[indices]
tsne = TSNE(n_components = 3, verbose = 2)
X_tsne = tsne.fit_transform(X)
```

### ITEM

```python
item=item_features.T
item.shape[0]
```

```python
from sklearn.manifold import TSNE
indices = np.random.choice(item.shape[0], 603, replace = False)
Y= item[indices]
tsne = TSNE(n_components = 3, verbose = 2)
Y_tsne = tsne.fit_transform(Y)
```

```python
from sklearn.cluster import KMeans
import collections
kmeans = KMeans(n_clusters=4, random_state=0).fit(X_tsne)
y_kmeans = kmeans.predict(X_tsne)
labels=kmeans.labels_

pickle.dump(labels, open( "pickles/labels.pkl", "wb" ), pickle.HIGHEST_PROTOCOL)
pickle.dump(X_tsne, open( "pickles/X_tsne.pkl", "wb" ), pickle.HIGHEST_PROTOCOL)
pickle.dump(Y_tsne, open( "pickles/Y_tsne.pkl", "wb" ), pickle.HIGHEST_PROTOCOL)
```

```python
indicesclust0= np.where( labels == 0 )
indicesclust1= np.where( labels == 1 )
indicesclust2= np.where( labels == 2 )
indicesclust3= np.where( labels == 3 )
cluster0=X_tsne[indicesclust0]
cluster1=X_tsne[indicesclust1]
cluster2=X_tsne[indicesclust2]
cluster3=X_tsne[indicesclust3]
```
