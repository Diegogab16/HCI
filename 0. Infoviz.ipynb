```python
def on_button_clickedX(widget,msg):
    #print(widget)
    #print(msg)

    curve=msg[0].get('curveNumber')
    number=msg[0].get('pointNumber')
    x_value=msg[0].get('x')
    y_value=msg[0].get('y')

    #print(curve)
    #print(msg[0].values())
    #graph.close()
    #graph1 = GraphWidget('https://plot.ly/~Diegogab/0')
    #display(graph1)
    if curve==0:
        show_curve0(number,x_value,y_value)

    if curve==1:
        show_curve1(number,x_value,y_value)
    if curve==2:
        show_curve2(number,x_value,y_value)
    if curve==3:
        show_curve3(number,x_value,y_value)


#graph.on_click(on_button_clickedX)

def show_curve0(numero, x,y):
    
    curve=0
    label_user=indicesclust0[0][numero]
    userpoint=X_tsne[label_user]
    initialmatrix=train.T.A
    rated_user=initialmatrix[label_user]
    indexes0=np.where(rated_user >= 4 )
    initial_items=Y_tsne[indexes0]
    final_rateduser=final_ratings[label_user] 
    indexes1=np.where(final_rateduser >= 4 )
    final_items=Y_tsne[indexes1]
    #print("Items recommended for User "+str(label_user))
    
    valid_moviesnames_init=valid_moviesnames[indexes0]
    valid_moviesgenre_init=valid_moviesgenre[indexes0]
    
    valid_moviesnames_fin=valid_moviesnames[indexes1]
    valid_moviesgenre_fin=valid_moviesgenre[indexes1]
    
    initial_info=[]
    final_info=[]
    
    for i in range(0,len(valid_moviesnames_init)):
        initial_info.append(valid_moviesnames_init[i]+", "+valid_moviesgenre_init[i])
    for j in range(0,len(valid_moviesgenre_fin)):
        final_info.append(valid_moviesnames_fin[j]+", "+valid_moviesgenre_fin[j])
    
    dibujar_graph(initial_items,final_items,label_user,initial_info,final_info)
    #global is_click
    #is_click=False
    
def show_curve1(numero, x,y):
    
    curve=1
    label_user=indicesclust1[0][numero]
    userpoint=X_tsne[label_user]
    initialmatrix=train.T.A
    rated_user=initialmatrix[label_user]
    indexes0=np.where(rated_user >= 4 )
    initial_items=Y_tsne[indexes0]
    final_rateduser=final_ratings[label_user] 
    indexes1=np.where(final_rateduser >= 4 )
    final_items=Y_tsne[indexes1]
    #print("Items recommended for User "+str(label_user))
    #initial_info=0
    #final_info=0
    valid_moviesnames_init=valid_moviesnames[indexes0]
    valid_moviesgenre_init=valid_moviesgenre[indexes0]
     
    valid_moviesnames_fin=valid_moviesnames[indexes1]
    valid_moviesgenre_fin=valid_moviesgenre[indexes1]
    
    initial_info=[]
    final_info=[]
    
    for i in range(0,len(valid_moviesnames_init)):
        initial_info.append(valid_moviesnames_init[i]+", "+valid_moviesgenre_init[i])
    for j in range(0,len(valid_moviesgenre_fin)):
        final_info.append(valid_moviesnames_fin[j]+", "+valid_moviesgenre_fin[j])
    
    dibujar_graph(initial_items,final_items,label_user,initial_info,final_info)
    #global is_click
    #is_click=False
    
def show_curve2(numero, x,y):
    
    curve=2
    label_user=indicesclust2[0][numero]
    userpoint=X_tsne[label_user]
    initialmatrix=train.T.A
    rated_user=initialmatrix[label_user]
    indexes0=np.where(rated_user >= 4 )
    initial_items=Y_tsne[indexes0]
    final_rateduser=final_ratings[label_user] 
    indexes1=np.where(final_rateduser >= 4 )
    final_items=Y_tsne[indexes1]
    #print("Items recommended for User "+str(label_user))
    #initial_info=0
    #final_info=0
    
    valid_moviesnames_init=valid_moviesnames[indexes0]
    valid_moviesgenre_init=valid_moviesgenre[indexes0]
    
    
    
    valid_moviesnames_fin=valid_moviesnames[indexes1]
    valid_moviesgenre_fin=valid_moviesgenre[indexes1]
    
    initial_info=[]
    final_info=[]
    
    for i in range(0,len(valid_moviesnames_init)):
        initial_info.append(valid_moviesnames_init[i]+", "+valid_moviesgenre_init[i])
    for j in range(0,len(valid_moviesgenre_fin)):
        final_info.append(valid_moviesnames_fin[j]+", "+valid_moviesgenre_fin[j])
    
    dibujar_graph(initial_items,final_items,label_user,initial_info,final_info)
    #global is_click
    #is_click=False
    
def show_curve3(numero, x,y):
    
    curve=3
    label_user=indicesclust3[0][numero]
    userpoint=X_tsne[label_user]
    initialmatrix=train.T.A
    rated_user=initialmatrix[label_user]
    indexes0=np.where(rated_user >= 4 )
    initial_items=Y_tsne[indexes0]
    final_rateduser=final_ratings[label_user] 
    indexes1=np.where(final_rateduser >= 4 )
    final_items=Y_tsne[indexes1]
    #print("Items recommended for User "+str(label_user))
    #initial_info=0
    #final_info=0
    
    valid_moviesnames_init=valid_moviesnames[indexes0]
    valid_moviesgenre_init=valid_moviesgenre[indexes0]
    
    
    
    valid_moviesnames_fin=valid_moviesnames[indexes1]
    valid_moviesgenre_fin=valid_moviesgenre[indexes1]
    
    initial_info=[]
    final_info=[]
    
    for i in range(0,len(valid_moviesnames_init)):
        initial_info.append(valid_moviesnames_init[i]+", "+valid_moviesgenre_init[i])
    for j in range(0,len(valid_moviesgenre_fin)):
        final_info.append(valid_moviesnames_fin[j]+", "+valid_moviesgenre_fin[j])
    
    dibujar_graph(initial_items,final_items,label_user,initial_info,final_info)
    #global is_click
    #is_click=False

def dibujar_graph(initial_items,final_items,label_user,initial_info,final_info): 
    trace0 = go.Scatter3d(
        x=initial_items[:, 0],
        y=initial_items[:, 1],
        z=initial_items[:, 2],
        name='Movies initially rated by user '+str(label_user),
        mode='markers',
        text=initial_info,
        marker=dict(
            #color='rgb(50, 50, 170)',
            size=15,
            symbol='circle',
            line=dict(
                #color='rgba(60,60,60)',
                width=1
            ),
            #opacity=0.8
        )
    )

    trace1 = go.Scatter3d(
        x=final_items[:, 0],
        y=final_items[:, 1],
        z=final_items[:, 2],
        name='Recommended Movies for user '+str(label_user),
        mode='markers',
        text=final_info,
        marker=dict(
            color='rgb(120, 240, 100)',
            size=12,
            symbol='circle',
            line=dict(
                #color='rgb(217,217,217)',
                width=1
            ),
            #opacity=0.9
        )
    )


    data = [trace0,trace1]
    layout = go.Layout(
        showlegend=True,
        title='Items for user '+str(label_user),
        margin=dict(
        l=0,
        r=0,
        b=0,
        t=-2
        ),
    )
    
    fig1 = go.Figure(data=data, layout=layout)
    #fig = Figure(data=data, layout=layout)
    #py.iplot(fig1, filename='Items')
    iplot(fig1)
def on_button_clicked5(b):
    global images24
    images24.on_click(on_button_clickedX)
    
    b.close()
    texto = widgets.HTML(tags.p("Now you can click on any of the users (Figure above) to get the recommended movies...").render())
    texto.layout.left='40px'
    display(texto)
    

def infoviz():
    global images24
    images24 = GraphWidget('https://plot.ly/~Diegogab/32/user-matrix/')
    texto = widgets.HTML(tags.p("Users are clustered in four groups. Click on the button 'View movies' to start...").render())
    texto.layout.left='40px'
    display(texto)
    update_display(images24,display_id='Prueba')
    
    #images24.on_click(on_button_clickedX)
    button5 = widgets.Button(description="View Movies",layout=Layout(width='50%', height='50px'))
    button5.style.button_color = 'lightsteelblue'
    button5.layout.left="20%"
    display(button5)
    button5.on_click(on_button_clicked5)
```
